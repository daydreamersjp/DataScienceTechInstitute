### Selecetion Sort
- Computation Complexity: O(n^2)
- Space Complexity: O(1)

```python
def selection_sort(ls):
    for i in range(len(ls)-1):
        smallest = i
        for j in range(i+1,len(ls)):
            if ls[j]<ls[smallest]: smallest = j
        ls[i],ls[smallest] = ls[smallest],ls[i]
    return ls
```

### Merge Sort
- A type of Divide and Conqure
- Computation Complexity: O(n * log(n))

```python
def merge_sort(ls):
    if len(ls)<=1: return ls
    if len(ls)==2: 
        if ls[0]<ls[1]: return ls
        else: return ls[::-1]
        
    mid = len(ls) // 2
    return merge(merge_sort(ls[:mid]), merge_sort(ls[mid:]))
```

## Merging Two Sorted List
- Computation Complexity: O(n)
- Space Complexity: O(n)

Tried two algos: one is using list and the other using stack (list but using as stack). Appearently, the latter is faster.

```python
ls1_s = [1,5,10]; ls2_s = [2,3,9]
ls12_b = [i for i in range(2*10**3)]
np.random.shuffle(ls12_b)
ls1_b = ls12_b[0:10**3]; ls2_b = ls12_b[10**3:2*10**3]
ls1_b.sort(); ls2_b.sort()
```

```python
del ls1_b
del ls2_b
del ls12_b
```

```python
def merge(ls1, ls2):
    res = []
    ls1r, ls2r = ls1[::-1], ls2[::-1]
    while (ls1r!=[] and ls2r!=[]):
        if (ls1r[-1]<ls2r[-1]):
            res.append(ls1r[-1])
            ls1r.pop()
        else:
            res.append(ls2r[-1])
            ls2r.pop()
            
    if (ls1r!=[]): res.extend(ls1r[::-1])
    if (ls2r!=[]): res.extend(ls2r[::-1])
        
    return res

def merge_nonstack(ls1, ls2):
    res = []
    ls1r, ls2r = [ls1[i] for i in range(len(ls1))], [ls2[i] for i in range(len(ls2))]
    while (ls1r!=[] and ls2r!=[]):
        if (ls1r[0]<ls2r[0]):
            res.append(ls1r[0])
            ls1r = ls1r[1:]
        else:
            res.append(ls2r[0])
            ls2r = ls2r[1:]
    
    if (ls1r!=[]): res.extend(ls1r)
    if (ls2r!=[]): res.extend(ls2r)    
    
        
    return res
```

## Generate Power Set
- Computation Complex: O(2^n)

```python
def power_set(ls):
    res = [[]]
    for item in ls:
        res.extend([subset + [item] for subset in res])
    return res
```

```python
%%time
sorted(power_set(['a','b','c']),key=len)
```

## Recursive
- Fibonacci sequence
    - Iterative is O(n), recursive without memorizing O(2^n)

```python
#Recursive: O(2^n)
def fib(n):
    if (isinstance(n,int)==False): return None
    if n<0: return None
    if n<=2: return 1
    return fib(n-1)+fib(n-2)

#Iterative: O(n)
def fib_ite(n):
    if (isinstance(n,int)==False): return None
    if n<0: return None
    if n<=2: return 1
    res = [1,1]
    for i in range(n-2):
        res.append(res[-2]+res[-1])
    return res[-1]

#Memorizing - Dynamic Programming
M = {1:1,2:1}
def fib_memo(n):
    if n not in M.keys():
        M[n] = fib_memo(n-1)+fib_memo(n-2)
    return M[n]
```

## Back Track

```python
class board:
    def __init__(self):
        self.state = [[0]*8 for _ in range(8)]
    
    def place_queen(self,position):
        self.state[position[0]][position[1]]=2
        self.attack_queen()
    
    def remove_queen(self,position):
        if self.state[position[0]][position[1]]!=2: pass
        self.state[position[0]][position[1]]=0
        self.attack_queen()
            
    def attack_queen(self):
        for i in range(len(self.state)):
            for j in range(len(self.state[0])):
                if self.state[i][j]!=2: self.state[i][j]=0
                    
        for i in range(len(self.state)):
            for j in range(len(self.state[0])):
                if self.state[i][j]==2:
                    upper = range(i-1,-1,-1)
                    lower = range(i+1,len(self.state),1)
                    left = range(j-1,-1,-1)
                    right = range(j+1,len(self.state[0]),1)
                    
                    if min(len(upper),len(left))>0:
                        for k in range(min(len(upper),len(left))): 
                            if self.state[upper[k]][left[k]]==0: self.state[upper[k]][left[k]]=1
                    if min(len(upper),len(right))>0:
                        for k in range(min(len(upper),len(right))): 
                            if self.state[upper[k]][right[k]]==0: self.state[upper[k]][right[k]]=1
                    if min(len(lower),len(left))>0:
                        for k in range(min(len(lower),len(left))): 
                            if self.state[lower[k]][left[k]]==0: self.state[lower[k]][left[k]]=1
                    if min(len(lower),len(right))>0:
                        for k in range(min(len(lower),len(right))): 
                            if self.state[lower[k]][right[k]]==0: self.state[lower[k]][right[k]]=1   
                                
                    if len(upper)>0:
                        for k in range(len(upper)):
                            if self.state[upper[k]][j]==0: self.state[upper[k]][j]=1
                    if len(lower)>0:
                        for k in range(len(lower)):
                            if self.state[lower[k]][j]==0: self.state[lower[k]][j]=1       
                    if len(left)>0:
                        for k in range(len(left)):
                            if self.state[i][left[k]]==0: self.state[i][left[k]]=1
                    if len(right)>0:
                        for k in range(len(right)):
                            if self.state[i][right[k]]==0: self.state[i][right[k]]=1   
    
    def count_queen(self):
        cnt = 0
        for i in range(len(self.state)):
            for j in range(len(self.state[0])):
                if self.state[i][j]==2: cnt += 1
        return cnt
```

```python
b = board()
```

```python
def queens(b):
    if b.count_queen()==8: 
        return True
    
    for i in range(len(b.state)):
        for j in range(len(b.state[0])):
            if b.state[i][j]==0:
                b.place_queen([i,j])
                if queens(b)==True: 
                    return True
                b.remove_queen([i,j])
    return False                
```

```python
def create_itemlist(num):
    names = [chr(97+i) for i in range(num)]
    values = np.random.uniform(5,50,num)
    weights = np.random.uniform(5,20,num)
    items = {}
    for i in range(len(names)):
        item = {}
        item['value'] = int(values[i])
        item['weight'] = int(weights[i])
        items[names[i]] = item
    return items
```

### Greedy
- Computation complexity: O(n)
- But not the global optimum, approximation approach

```python
def greedy_knapsack(items,max_weight):
    bag_weight = 0
    bag_items = []
    for item in sorted(items, key=lambda x: (items[x]['value']), reverse=True):
        if max_weight >= bag_weight+items[item]['weight']:
            bag_weight += items[item]['weight']
            bag_items += [item]
    return sorted(bag_items)
```

```python
%%time
res = greedy_knapsack(items,80)
print('Items in knapsack:',res)
print('Final value:',sum(items[x]['value'] for x in res))
print('Final weight:',sum(items[x]['weight'] for x in res))
```

### Divide and Conquer - Recursive
- Computation Complexity: O(nW)
- Divide problem to two: 1. a problem when choosing an item A, 2. a problem when removing an item A from the list.

```python
def dc_knapsack2(items,max_weight): #Return: total values, total weights, selected items
    if len(items)==0: 
        return 0,0,[]
    if max_weight < min([items[x]['weight'] for x in items.keys()]):
        return 0,0,[]
    items_removed = {k:v for k,v in items.items()}
    items_removed.pop(list(items.keys())[0])
    
    v1, w1, it1 = dc_knapsack2(items_removed,max_weight)
    v2, w2, it2 = dc_knapsack2(items_removed,max_weight-items[list(items.keys())[0]]['weight'])
    v2 += items[list(items.keys())[0]]['value']
    w2 += items[list(items.keys())[0]]['weight']
    it2 += [list(items.keys())[0]]
    
    if v1>v2:
        return v1,w1,sorted(it1)
    return v2,w2,sorted(it2)
```

### Branch and Bound
- Lower bound: Greedy algorithm
- Upper bound: Allow partial items packing -> Pack from higher value-weight ratio

```python
def greedy_knapsack_bb(items,max_weight): #Return: total values, total weights, selected items
    
    bag_value = 0
    bag_weight = 0
    bag_items = []
    for item in sorted(items, key=lambda x: (items[x]['value']), reverse=True):
        if max_weight >= bag_weight+items[item]['weight']:
            bag_value += items[item]['value']
            bag_weight += items[item]['weight']
            bag_items += [item]
    return bag_value, bag_weight, sorted(bag_items)
```

```python
def powdered_knapsack_bb(items,max_weight): 
    
    bag_value = 0
    bag_weight = 0
    bag_items = []
    for item in sorted(items, key=lambda x: items[x]['value']/items[x]['weight'], reverse=True):
        if max_weight >= bag_weight+items[item]['weight']:
            bag_value += items[item]['value']
            bag_weight += items[item]['weight']
            bag_items += [item]
        elif (max_weight > bag_weight and max_weight < bag_weight+items[item]['weight']):
            bag_value += items[item]['value']/items[item]['weight'] * (max_weight - bag_weight)
            bag_weight = max_weight
            bag_items += [item]
    return bag_value, bag_weight, sorted(bag_items)
```

```python
def bb_knapsack(items,max_weight): #Return: total values, total weights, selected items
    global glob_lower
    global None_cnt
    global Val_cnt
    
    if len(items)==0: 
        return 0,0,[],0,0
    if max_weight < min([items[x]['weight'] for x in items.keys()]):
        return 0,0,[],0,0
    
    glob_lower, _, _ = greedy_knapsack_bb(items,max_weight)
    
    items_removed = {k:v for k,v in items.items()}
    items_removed.pop(list(items.keys())[0])
    
    upper1, _, _ = powdered_knapsack_bb(items_removed,max_weight)
    upper2, _, _ = powdered_knapsack_bb(items_removed,max_weight-items[list(items.keys())[0]]['weight'])
    upper2 += items[list(items.keys())[0]]['value']
    
    if glob_lower <= upper1:
        v1, w1, it1, _, _ = bb_knapsack(items_removed,max_weight)
        v1l,_,_ = greedy_knapsack_bb(items_removed,max_weight)
        Val_cnt += 1
    else:
        v1, w1, it1 = None,None,None
        v1l = None
        None_cnt += 1
    
    if glob_lower <= upper2:
        v2, w2, it2, _, _ = bb_knapsack(items_removed,max_weight-items[list(items.keys())[0]]['weight'])
        v2 += items[list(items.keys())[0]]['value']
        w2 += items[list(items.keys())[0]]['weight']
        it2 += [list(items.keys())[0]]   
        v2l,_,_ = greedy_knapsack_bb(items_removed,max_weight-items[list(items.keys())[0]]['weight'])
        v2l += items[list(items.keys())[0]]['value']
        Val_cnt += 1
    else:
        v2, w2, it2 = None,None,None
        v2l = None
        None_cnt += 1
    
    if (v1==None):
        if glob_lower < v2l: 
            glob_lower = v2l
        return v2,w2,sorted(it2),Val_cnt,None_cnt
    elif (v2==None):
        if glob_lower < v1l: 
            glob_lower = v1l
        return v1,w1,sorted(it1),Val_cnt,None_cnt
    else:
        if glob_lower < min(v1l,v2l): 
            glob_lower = min(v1l,v2l)        
        if v1>v2:
            return v1,w1,sorted(it1),Val_cnt,None_cnt
        return v2,w2,sorted(it2),Val_cnt,None_cnt
```
